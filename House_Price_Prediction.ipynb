```python
for x in tdf.columns:
    if tdf[x].dtype != 'object':
        Q1 = tdf[x].quantile(0.25)
        Q3 = tdf[x].quantile(0.75)
        IQR = Q3 - Q1
        lb = Q1 - 1.5 * IQR
        ub = Q3 + 1.5 * IQR
        tdf = tdf[(tdf[x]>= lb) & (tdf[x] <= ub)]
```

```python
X_train = tdf.iloc[:,:-1]
y_train = tdf['SalePrice']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(n_estimators = 500, learning_rate = 0.05, max_depth = 5, subsample = 0.8, random_state = 20)
```

```python
gbr.fit(X_train, y_train)
```

```python
tstdf = tstdf.drop(columns = 'SalePrice')
```

```python
X_test = tstdf
X_test
```

```python
y_pred = gbr.predict(X_test)
```

```python
# Get actual prices (from training)
actual = tdf['SalePrice'].values  # shape: (n_train,)
preds = y_pred                 # shape: (n_test,)

# Combine: actual first, then predictions
combined = np.concatenate([actual, preds])

# Create x-axis: 0 to n_train + n_test
x = np.arange(len(combined))

plt.figure(figsize=(12, 6))
plt.plot(x[:len(actual)], actual, label='Actual (Train)', color='blue')
plt.plot(x[len(actual):], preds, label='Predicted (Test)', color='red', linestyle='--')
plt.axvline(x=len(actual)-1, color='gray', linestyle=':', label='Train/Test Split')
plt.legend()
plt.title('House Prices: Actual vs Predicted (Test as Continuation)')
plt.xlabel('Sample Index')
plt.ylabel('SalePrice')
plt.show()
```
